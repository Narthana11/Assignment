```python
#A2.2 : Motion Logging

#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
#Reading the CSV File
df = pd.read_csv("AccDataRP2040indexed.csv")
#Printing the DataFrame
print(df)
```

```python


time_values = pd.to_datetime(df['Time']).sub(pd.to_datetime(df['Time'].iloc[0])).dt.total_seconds()


fig, ax = plt.subplots(figsize=(10, 6))


ax.plot(time_values, df['Ax'], label='Ax', color='blue')
ax.plot(time_values, df['Ay'], label='Ay', color='red')
ax.plot(time_values, df['Az'], label='Az', color='green')
ax.plot(time_values, df['Gx'], label='Gx', color='orange')
ax.plot(time_values, df['Gy'], label='Gy', color='purple')
ax.plot(time_values, df['Gz'], label='Gz', color='brown')

# Set labels and title
ax.set_xlabel('Time (seconds)')
ax.set_ylabel('Acceleration (m/sÂ²) / Gyroscope (rad/s)')  # Assuming mixed units
ax.set_title('Motion Data from AccDataRP2040indexed.csv')


ax.grid(True)


ax.legend(loc='upper right')


plt.show()
```

```python
ax_threshold = 0.1
ay_threshold = 0.1
az_threshold = 0.1

# Filter the DataFrame based on the thresholds
filtered_df = df[
    (df['Ax'].abs() > ax_threshold) |
    (df['Ay'].abs() > ay_threshold) |
    (df['Az'].abs() > az_threshold)
]


time_values = pd.to_datetime(filtered_df['Time']).sub(pd.to_datetime(filtered_df['Time'].iloc[0])).dt.total_seconds()


fig, ax = plt.subplots(figsize=(10, 6))


ax.plot(time_values, filtered_df['Ax'], label='Ax', color='blue')
ax.plot(time_values, filtered_df['Ay'], label='Ay', color='red')
ax.plot(time_values, filtered_df['Az'], label='Az', color='green')

ax.set_xlabel('Time (seconds)')
ax.set_ylabel('Acceleration (m/sÂ²)')
ax.set_title('Compressed Motion Data (Acceleration Only)')


ax.grid(True)


ax.legend(loc='upper right')


plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = pd.read_csv('Climate2016.csv')

data['windvelo_x'] = data['windvelo (m/s)'] * np.cos(np.radians(data['winddeg (deg)']))
data['windvelo_y'] = data['windvelo (m/s)'] * np.sin(np.radians(data['winddeg (deg)']))


def normalize(series):
    return (series - series.min()) / (series.max() - series.min())


data['windvelo_x_norm'] = normalize(data['windvelo_x'])
data['windvelo_y_norm'] = normalize(data['windvelo_y'])


plt.figure(figsize=(10, 8))
plt.hist2d(data['winddeg (deg)'], data['windvelo (m/s)'], bins=(50, 50), vmax=400)
plt.colorbar(label='Frequency')
plt.xlabel('Wind Direction (degrees)')
plt.ylabel('Wind Speed (m/s)')
plt.title('Original Wind Data Distribution')
plt.show()


plt.figure(figsize=(10, 8))
plt.hist2d(data['windvelo_x'], data['windvelo_y'], bins=(50, 50), vmax=400)
plt.colorbar(label='Frequency')
plt.xlabel('Wind Velocity X Component (m/s)')
plt.ylabel('Wind Velocity Y Component (m/s)')
plt.title('Transformed Wind Vector Distribution')
plt.show()

# Save the updated data to a new CSV file
data.to_csv('Climate2016_with_vectors.csv', index=False)

```
